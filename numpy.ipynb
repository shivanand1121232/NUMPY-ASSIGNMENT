```python
#3. Describe the methods for reversing a NumPy array along different axes. Provide examples for 1D and 2D
#arrays.


NumPy provides several methods to reverse an array along different axes:

1. np.flip(): Reverses the array along the specified axis.

2. np.flipud(): Reverses the array along the vertical axis (axis 0).

3. np.fliplr(): Reverses the array along the horizontal axis (axis 1).

4. array[::-1]: Reverses the entire array.

Examples:

1D Array:


import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(np.flip(arr))  # Output: [5 4 3 2 1]
print(arr[::-1])  # Output: [5 4 3 2 1]


2D Array:


arr = np.array([[1, 2, 3], [4, 5, 6]])
print(np.flip(arr, axis=0))  # Output: [[4 5 6], [1 2 3]]
print(np.flip(arr, axis=1))  # Output: [[3 2 1], [6 5 4]]
print(np.flipud(arr))  # Output: [[4 5 6], [1 2 3]]
print(np.fliplr(arr))  # Output: [[3 2 1], [6 5 4]]


In the 2D array examples:

- axis=0 reverses the rows (vertical axis).
- axis=1 reverses the columns (horizontal axis).
- np.flipud() reverses the rows.
- np.fliplr() reverses the columns.
```

```python
#7. Compare vstack() and hstack() functions in NumPy. Provide examples demonstrating their usage and
#output.


NumPy's vstack() and hstack() functions are used to stack arrays vertically and horizontally, respectively.

vstack()

- Stacks arrays vertically (row-wise)
- Increases the number of rows
- Example:


import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.vstack((a, b)))
# Output:
# [[1 2 3]
#  [4 5 6]]


hstack()

- Stacks arrays horizontally (column-wise)
- Increases the number of columns
- Example:


import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.hstack((a, b)))
# Output:
# [1 2 3 4 5 6]
```

```python
#8. Explain the differences between fliplr() and flipud() methods in NumPy, including their effects on various
#array dimensions.

NumPy's fliplr() and flipud() methods are used to flip arrays horizontally (left-right) and vertically (up-down), respectively.

fliplr():

- Flips the array horizontally (left-right)
- Reverses the order of columns
- Does not change the number of rows

Example:


import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])

print(np.fliplr(arr))
# Output:
# [[3 2 1]
#  [6 5 4]]


flipud():

- Flips the array vertically (up-down)
- Reverses the order of rows
- Does not change the number of columns

Example:


import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])

print(np.flipud(arr))
# Output:
# [[4 5 6]
#  [1 2 3]]


Effects on various array dimensions:

- 1D array: fliplr() is equivalent to flipud(), as there's only one dimension to flip.
- 2D array:
    - fliplr() flips columns.
    - flipud() flips rows.
- 3D array or higher:
    - fliplr() flips the last dimension (columns).
    - flipud() flips the first dimension (rows)
```

```python
#9. Discuss the functionality of the array_split() method in NumPy. How does it handle uneven splits?



The array_split() method in NumPy splits an array into multiple sub-arrays along a specified axis. It takes three arguments:

1. ary: The input array to be split.
2. indices_or_sections: The number of splits or the indices at which to split.
3. axis: The axis along which to split (default is 0).

If indices_or_sections is an integer, the array is split into equal-sized sub-arrays. 
If it's a list of integers, the array is split at those specific indices.

Handling uneven splits:

- If the array cannot be evenly split, array_split() will adjust the size of the 
last sub-array to accommodate the remaining elements.
- If indices_or_sections is a list of integers, the splits will be made at those exact indices, regardless 
of whether the splits are even or uneven.
```

```python
#10. Explain the concepts of vectorization and broadcasting in NumPy. How do they contribute to efficient array
#operations?



Vectorization and broadcasting are two fundamental concepts in NumPy that enable efficient array operations.

Vectorization:

Vectorization refers to the ability to perform operations on entire arrays at once, without the need for explicit loops. This is achieved by using NumPy's universal functions (ufuncs), which are designed to work with arrays. Vectorization allows for:

- Element-wise operations
- Matrix multiplication
- Array-wide reductions (e.g., sum, mean)

Example:


import numpy as np

arr = np.array([1, 2, 3, 4, 5])
double_arr = arr * 2  # Vectorized operation
print(double_arr)  # [2 4 6 8 10]


Broadcasting:

Broadcasting is the process of aligning arrays with different shapes to enable element-wise operations. NumPy follows these broadcasting rules:

- If arrays have the same shape, no broadcasting is needed.
- If arrays have different shapes, NumPy tries to broadcast them by:
    - Adding dimensions of size 1 to the smaller array.
    - Repeating the smaller array along the new dimension.

Example:


import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Broadcasting: arr1 is repeated along the new dimension
result = arr1 + arr2[:, np.newaxis]
print(result)  # [[5 6 7], [6 7 8], [7 8 9]]


Contribution to efficient array operations:

Vectorization and broadcasting enable efficient array operations by:

- Avoiding explicit loops, which reduces overhead and improves performance.
- Allowing for parallelization, as operations can be executed simultaneously on multiple elements.
- Enabling the use of optimized, low-level code, which further improves performance.

By leveraging vectorization and broad
```
